```
class NodoArbol:
  def __init__(self, value, left=None, right=None):
    self.data=value
    self.left=left
    self.right=right
```

```
class BinarySearchTree:

  def __init__(self):
    self.__root=None

  def insert(self, value):
    if self.__root==None:
       self.__root= NodoArbol(value,None,None)
    else:
      #preguntar si value es menor que root, de ser el caso
      #insertar a la izq, pero puede ser el caso de que el 
      #sub arbol izq ya tenga muchos elementos
      self._insert_nodo_(self.__root,value)
  
  def _insert_nodo_(self,nodo,value):
    if nodo.data==value:
      pass
    elif value<nodo.data: #true va a la izq
      if nodo.left==None: #si hay espacio en la izq, ahi va
        nodo.left=NodoArbol(value,None,None)#insertamos el nodo
      else: 
        self._insert_nodo_(nodo.left,value)#buscar el sub arbol izq
    else:
      if nodo.right==None:
        nodo.right=NodoArbol(value,None,None)
      else:
        self._insert_nodo_(nodo.right,value)##buscar en sub arbol der
  
  def buscar(self, value):
     if self.__root==None:
       return None
     else:
       #haremos busqueda recursiva
       return self.__busca_nodo(self.__root,value)
  
  def __busca_nodo(self,nodo,value):
     if nodo ==None:
       return False
     elif nodo.data==value:
       return nodo.data
     elif value< nodo.data:
       return self.__busca_nodo(nodo.left,value)
     else:
       return self.__busca_nodo(nodo.right,value) 
  
  def transversal(self,format="inorden"):
    if format =="inorden":
      self.__recorrido_in(self.__root)
    elif format=="preorden":
      self.__recorrido_pre(self.__root)
    elif format =="posorden":
      self.__recorrido_pos(self.__root)
    else:
      print("Fotmato de recorrido no valido")
  
  def __recorrido_pre(self, nodo):
    if nodo != None:
      print(nodo.data, end=",")
      self.__recorrido_pre(nodo.left)
      self.__recorrido_pre(nodo.right)
  
  def __recorrido_in(self, nodo):
    if nodo != None:
      self.__recorrido_in(nodo.left)
      print(nodo.data, end=",")
      self.__recorrido_in(nodo.right)
  
  def __recorrido_pos(self, nodo):
    if nodo!= None:
      self.__recorrido_pos(nodo.left)
      self.__recorrido_pos(nodo.right)
      print(nodo.data, end=",")

  def eliminar(self, value):
    self.__eliminar_nodo(self.__root, self.__root, self.__root, None, value)

  def __eliminar_nodo(self, root, nodo, ant, actual, value):
    if nodo==None:
      return print("Error de nodo")
    if nodo.data==value:
      if nodo.left==None and nodo.right==None: 
        if actual=="izq":
          ant.left=None
        elif actual=="der":
          ant.right=None
        print(f"Nodo eiminado: {value}")
      elif nodo.left==None and nodo.right!=None:
        if actual=="izq":
          ant.left=nodo.right
        else:
          ant.right=nodo.right
        print(f"Nodo eiminado: {value}")
      elif nodo.left!=None and nodo.right==None:
        if actual=="izq":
          ant.left=nodo.left
        else:
          ant.right=nodo.left
        print(f"Nodo eiminado: {value}")
      elif nodo.left!=None and nodo.right!=None: 
        tmp,ant2 =self.nodoMasBajo(nodo.right, nodo)
        if nodo.data==ant2.data:
          ant2.right=tmp.right
        elif nodo.data!=ant2.data:
          ant2.left=tmp.right
        if actual=="izq":
          ant.left=tmp
        else:
          ant.right=tmp
        tmp.left=nodo.left
        tmp.right=nodo.right
        print(f"Nodo eiminado: {value}")
    elif value<nodo.data:
      return self.__eliminar_nodo(root, nodo.left, nodo, "izq", value)
    else:
      return self.__eliminar_nodo(root, nodo.right, nodo, "der", value)
    
  def nodoMasBajo(self, nodo, anterior):
    if nodo.left==None:
      return nodo, anterior
    elif nodo.left!=None:
      return self.nodoMasBajo(nodo.left, nodo)
```
